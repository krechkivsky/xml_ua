````python
# –í–∏–¥–∞–ª–µ–Ω–Ω—è –∫–æ–º–µ–Ω—Ç–∞—Ä—ñ–≤
import os
import re
import shutil
import datetime
import subprocess

def create_backup(backup_dir):
    """–°—Ç–≤–æ—Ä—é—î —Ä–µ–∑–µ—Ä–≤–Ω—É –∫–æ–ø—ñ—é —Ñ–∞–π–ª—ñ–≤ —É –ø–æ—Ç–æ—á–Ω–æ–º—É –∫–∞—Ç–∞–ª–æ–∑—ñ."""

    now = datetime.datetime.now()
    date_time = now.strftime("%Y-%m-%d %H-%M")

    backup_path = os.path.join(backup_dir, date_time)

    if not os.path.exists(backup_dir):
        os.makedirs(backup_dir)

    if os.path.exists(backup_path):
        print(f"–ü–æ–º–∏–ª–∫–∞: –î–∏—Ä–µ–∫—Ç–æ—Ä—ñ—è '{backup_path}' –≤–∂–µ —ñ—Å–Ω—É—î.")
        return False

    os.makedirs(backup_path)
    print(f"–°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ä–µ–∑–µ—Ä–≤–Ω–æ—ó –∫–æ–ø—ñ—ó –≤ '{backup_path}'...")

    # –ö–æ–ø—ñ—é–≤–∞–Ω–Ω—è —Ñ–∞–π–ª—ñ–≤
    for ext in ["*.py", "*.ini", "*.ui", "*.bat"]:
        for filename in os.listdir("."):
            if filename.endswith(ext[1:]):
                src = os.path.join(".", filename)
                dst = os.path.join(backup_path, filename)
                shutil.copy2(src, dst)  # copy2 –∑–±–µ—Ä—ñ–≥–∞—î –º–µ—Ç–∞–¥–∞–Ω—ñ
                print(f"  –°–∫–æ–ø—ñ–π–æ–≤–∞–Ω–æ: {filename}")

    # –ö–æ–ø—ñ—é–≤–∞–Ω–Ω—è –ø–∞–ø–æ–∫ –∑ –≤–º—ñ—Å—Ç–æ–º
    for folder in ["help", "docs", "images", "templates"]:
        src_folder = os.path.join(".", folder)
        dst_folder = os.path.join(backup_path, folder)
        if os.path.exists(src_folder):
            shutil.copytree(src_folder, dst_folder, dirs_exist_ok=True)
            print(f"  –°–∫–æ–ø—ñ–π–æ–≤–∞–Ω–æ –ø–∞–ø–∫—É: {folder}")

    print(f"–†–µ–∑–µ—Ä–≤–Ω–∞ –∫–æ–ø—ñ—è —Å—Ç–≤–æ—Ä–µ–Ω–∞ –≤ '{backup_path}'.")
    return True


def remove_working_comments(directory):
    """–í–∏–¥–∞–ª—è—î —Ä–æ–±–æ—á—ñ –∫–æ–º–µ–Ω—Ç–∞—Ä—ñ (#) –∑ —Ñ–∞–π–ª—ñ–≤ *.py –≤ –∑–∞–¥–∞–Ω—ñ–π –¥–∏—Ä–µ–∫—Ç–æ—Ä—ñ—ó."""
    for filename in os.listdir(directory):
        if filename.endswith(".py"):
            filepath = os.path.join(directory, filename)
            try:
                with open(filepath, "r", encoding="utf-8") as f:
                    lines = f.readlines()

                new_lines = []
                for line in lines:
                    # –í–∏–¥–∞–ª—è—î–º–æ –∫–æ–º–µ–Ω—Ç–∞—Ä—ñ, —è–∫—ñ –ø–æ—á–∏–Ω–∞—é—Ç—å—Å—è –∑ '#' –Ω–∞ –ø–æ—á–∞—Ç–∫—É —Ä—è–¥–∫–∞ –∞–±–æ –ø—ñ—Å–ª—è –ø—Ä–æ–±—ñ–ª—ñ–≤
                    # –∞–ª–µ –∑–∞–ª–∏—à–∞—î–º–æ –∫–æ–º–µ–Ω—Ç–∞—Ä—ñ –≤ —Å–µ—Ä–µ–¥–∏–Ω—ñ —Ä—è–¥–∫–∞
                    line = re.sub(r"^(#+)(.*)$", "", line)
                    line = re.sub(r"^\s+(#+)(.*)$", "", line)
                    new_lines.append(line)

                with open(filepath, "w", encoding="utf-8") as f:
                    f.writelines(new_lines)

                print(f"–ö–æ–º–µ–Ω—Ç–∞—Ä—ñ –≤–∏–¥–∞–ª–µ–Ω–æ –∑ —Ñ–∞–π–ª—É: {filename}")

            except Exception as e:
                print(f"–ü–æ–º–∏–ª–∫–∞ –ø—Ä–∏ –æ–±—Ä–æ–±—Ü—ñ —Ñ–∞–π–ª—É {filename}: {e}")


def main():
    """–û—Å–Ω–æ–≤–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è —Å–∫—Ä–∏–ø—Ç—É."""

    answer = input(
        "–í–∏ –¥—ñ–π—Å–Ω–æ —Ö–æ—á–µ—Ç–µ –≤–∏–¥–∞–ª–∏—Ç–∏ —Ä–æ–±–æ—á—ñ –∫–æ–º–µ–Ω—Ç–∞—Ä—ñ –∑ —Ñ–∞–π–ª—ñ–≤ *.py? (y/n): "
    ).lower()
    result = answer == "y"

    if result:
        backup_dir = "./backups"
        if create_backup(backup_dir):
            remove_working_comments(".")
        else:
            print("–ü–æ–º–∏–ª–∫–∞: –ù–µ –≤–¥–∞–ª–æ—Å—è —Å—Ç–≤–æ—Ä–∏—Ç–∏ —Ä–µ–∑–µ—Ä–≤–Ω—É –∫–æ–ø—ñ—é.")
    else:
        print("–í–∏–¥–∞–ª–µ–Ω–Ω—è –∫–æ–º–µ–Ω—Ç–∞—Ä—ñ–≤ —Å–∫–∞—Å–æ–≤–∞–Ω–æ.")


if __name__ == "__main__":
    main()
````

````python
# –í—ñ–¥–Ω–æ–≤–ª–µ–Ω–Ω—è
import os
import shutil
import glob

def restore_backup(backup_dir="./backups"):
    """
    –í—ñ–¥–Ω–æ–≤–ª—é—î —Ñ–∞–π–ª–∏ –∑ –æ—Å—Ç–∞–Ω–Ω—å–æ—ó —Ä–µ–∑–µ—Ä–≤–Ω–æ—ó –∫–æ–ø—ñ—ó, –≤–∏–º–∞–≥–∞—é—á–∏ —è–≤–Ω–æ–≥–æ –ø—ñ–¥—Ç–≤–µ—Ä–¥–∂–µ–Ω–Ω—è.

    Args:
        backup_dir (str): –®–ª—è—Ö –¥–æ –¥–∏—Ä–µ–∫—Ç–æ—Ä—ñ—ó, –¥–µ –∑–±–µ—Ä—ñ–≥–∞—é—Ç—å—Å—è —Ä–µ–∑–µ—Ä–≤–Ω—ñ –∫–æ–ø—ñ—ó.
                          –ó–∞ –∑–∞–º–æ–≤—á—É–≤–∞–Ω–Ω—è–º "./backups".
    """

    print("–£–í–ê–ì–ê! –¶—è –æ–ø–µ—Ä–∞—Ü—ñ—è –≤—ñ–¥–Ω–æ–≤–∏—Ç—å —Ñ–∞–π–ª–∏ –∑ –æ—Å—Ç–∞–Ω–Ω—å–æ—ó —Ä–µ–∑–µ—Ä–≤–Ω–æ—ó –∫–æ–ø—ñ—ó.")
    print("–¶–µ –º–æ–∂–µ –ø—Ä–∏–∑–≤–µ—Å—Ç–∏ –¥–æ –≤—Ç—Ä–∞—Ç–∏ –ø–æ—Ç–æ—á–Ω–∏—Ö –∑–º—ñ–Ω. –ë—É–¥—å—Ç–µ –æ–±–µ—Ä–µ–∂–Ω—ñ!")
    confirmation = input("–í–≤–µ–¥—ñ—Ç—å 'yes' –¥–ª—è –ø—ñ–¥—Ç–≤–µ—Ä–¥–∂–µ–Ω–Ω—è –≤—ñ–¥–Ω–æ–≤–ª–µ–Ω–Ω—è: ")

    if confirmation.lower() != "yes":
        print("–í—ñ–¥–Ω–æ–≤–ª–µ–Ω–Ω—è —Å–∫–∞—Å–æ–≤–∞–Ω–æ.")
        return False

    print("–í—ñ–¥–Ω–æ–≤–ª–µ–Ω–Ω—è —Ñ–∞–π–ª—ñ–≤...")

    # –ó–Ω–∞—Ö–æ–¥–∏–º–æ –æ—Å—Ç–∞–Ω–Ω—é —Ä–µ–∑–µ—Ä–≤–Ω—É –∫–æ–ø—ñ—é
    backup_folders = glob.glob(os.path.join(backup_dir, "*"))
    if not backup_folders:
        print(f"–ü–æ–º–∏–ª–∫–∞: –ù–µ –∑–Ω–∞–π–¥–µ–Ω–æ –∂–æ–¥–Ω–æ—ó —Ä–µ–∑–µ—Ä–≤–Ω–æ—ó –∫–æ–ø—ñ—ó –≤ '{backup_dir}'.")
        return False

    latest_backup = max(backup_folders, key=os.path.getctime)

    print(f"–í—ñ–¥–Ω–æ–≤–ª–µ–Ω–Ω—è –∑ '{latest_backup}'...")

    # –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ, —á–∏ —ñ—Å–Ω—É—î –¥–∏—Ä–µ–∫—Ç–æ—Ä—ñ—è —Ä–µ–∑–µ—Ä–≤–Ω–æ—ó –∫–æ–ø—ñ—ó
    if not os.path.exists(latest_backup):
        print(f"–ü–æ–º–∏–ª–∫–∞: –ö–∞—Ç–∞–ª–æ–≥ '{latest_backup}' –Ω–µ —ñ—Å–Ω—É—î.")
        return False

    # –ö–æ–ø—ñ—é–≤–∞–Ω–Ω—è —Ñ–∞–π–ª—ñ–≤
    for ext in ["*.py", "*.ini", "*.ui", "*.bat"]:
        for filename in glob.glob(os.path.join(latest_backup, ext)):
            try:
                shutil.copy2(filename, ".")
                print(f"  –í—ñ–¥–Ω–æ–≤–ª–µ–Ω–æ: {os.path.basename(filename)}")
            except Exception as e:
                print(f"–ü–æ–º–∏–ª–∫–∞ –ø—Ä–∏ –≤—ñ–¥–Ω–æ–≤–ª–µ–Ω–Ω—ñ {filename}: {e}")

    # –í—ñ–¥–Ω–æ–≤–ª–µ–Ω–Ω—è –ø–∞–ø–æ–∫ –∑ –≤–º—ñ—Å—Ç–æ–º
    folders_to_restore = ["help", "docs", "images", "templates"]
    for folder in folders_to_restore:
        src_folder = os.path.join(latest_backup, folder)
        dst_folder = os.path.join(".", folder)
        if os.path.exists(src_folder):
            try:
                if os.path.exists(dst_folder):
                    shutil.rmtree(dst_folder)
                shutil.copytree(src_folder, dst_folder, dirs_exist_ok=False)
                print(f"  –í—ñ–¥–Ω–æ–≤–ª–µ–Ω–æ –ø–∞–ø–∫—É: {folder}")
            except Exception as e:
                print(f"–ü–æ–º–∏–ª–∫–∞ –ø—Ä–∏ –≤—ñ–¥–Ω–æ–≤–ª–µ–Ω–Ω—ñ –ø–∞–ø–∫–∏ {folder}: {e}")

    print("–í—ñ–¥–Ω–æ–≤–ª–µ–Ω–Ω—è –∑–∞–≤–µ—Ä—à–µ–Ω–æ.")
    return True


if __name__ == "__main__":
    restore_backup()
````
